```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "Resources/chopped_dataset2015.csv"
data2 = "Resources/population.csv"
```

```python
yearly = pd.read_csv(data)
yearly_df = pd.DataFrame(yearly)
yearly_df.head()
```

```python
grouped_yearly_df = yearly_df.groupby(['country_code'])

grouped_yearly_df.count().head(10)
```

```python
ConstantAmount_df = pd.DataFrame(
    grouped_yearly_df["constant_amount"].sum())

ConstantAmount_df.head(10)
```

```python
population = pd.read_csv(data2)
df = pd.DataFrame(population)
df.head()
```

```python
df_filtered = df[['country_code', 'Country Name', '2015']]

df_filtered
```

```python
merged_df = pd.merge(df_filtered, ConstantAmount_df, how = 'left', 
                     on=['country_code'])
merged_df
```

```python
merged_df = merged_df.rename(columns= {'country_code': 'Country Code', 'constant_amount': '2015 AID' })
```

```python
merged_df.to_csv('2015_AID.csv')
```
